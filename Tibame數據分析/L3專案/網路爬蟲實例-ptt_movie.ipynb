```python
import requests
from bs4 import BeautifulSoup
```

```python
#利用網址規律換頁
page = 10224
while page >= 10220:
    print("第{}頁".format(page))
    
    url = "https://www.ptt.cc/bbs/movie/index{}.html".format(page)
    headers = {'User-Agent' :'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'}

    res = requests.get(url, headers = headers)

    soup = BeautifulSoup(res.text, 'html.parser')

    #print(soup.prettify())

    article_title_html = soup.select("div[class='title']")
    #print(article_title_html)
    
    for article in article_title_html:
        try:
            print(article.a.text)
            print('https://www.ptt.cc' + article.a["href"])
        except:
            continue
    page -= 1
    
```

```python
#利用「上一頁」標籤位子換頁
url = "https://www.ptt.cc/bbs/movie/index.html"
headers = {'User-Agent' :'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'}

for i in range(0,3):

    res = requests.get(url, headers = headers)

    soup = BeautifulSoup(res.text, 'html.parser')

    article_title_html = soup.select("div[class='title']")

    for article in article_title_html:
        try:
            print(article.a.text)
            print('https://www.ptt.cc' + article.a["href"])
        except:
            continue
    
    #上一頁的標籤
    last_page_url = soup.select("a[class='btn wide']")[1]["href"]
    last_page_url = 'https://www.ptt.cc' + last_page_url
    url = last_page_url
```

```python
#儲存內文
import os

# 在當前路徑下創建res資料夾
resource_path = './res'
if not os.path.exists(resource_path):
    os.mkdir(resource_path)

url = "https://www.ptt.cc/bbs/movie/index.html"
headers = {'User-Agent' :'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'}

for i in range(0,2):

    res = requests.get(url, headers = headers)

    soup = BeautifulSoup(res.text, 'html.parser')

    article_title_html = soup.select("div[class='title']")

    for article in article_title_html:
        try:
            article_text = article.a.text
            article_url = 'https://www.ptt.cc' + article.a["href"]
            
            print(article_text)
            print(article_url)
            
            article_text = article_text.replace(":", "").replace("/", "").replace("*", "").replace("?", "").replace("\"", "").replace("<", "").replace(">", "").replace("|", "").replace(" ", "_").strip()
            #對文章網址請求
            article_res = requests.get(article_url, headers = headers)
            article_soup = BeautifulSoup(article_res.text, 'html.parser')
            
            article_content = article_soup.select("div#main-content")[0].text.split('--')[0]
            
            print(article_content)
            
            with open('{}/{}.txt'.format(resource_path, article_text), 'w', encoding='utf-8') as file:
                file.write(article_content)
        except:
            continue
    
    #上一頁的標籤
    last_page_url = soup.select("a[class='btn wide']")[1]["href"]
    last_page_url = 'https://www.ptt.cc' + last_page_url
    url = last_page_url
```
